```python
# task1
def reverseLookUp(s, v):
    keys = []
    for i,j in s.items():
        if j == v: 
            keys.append(i)
    return keys
dict1 = {'Ayat':18, 'Amir':17, 'Yergeldi':18,'Almas':22}
print(reverseLookUp(dict1, 16))
print(reverseLookUp(dict1, 17))
print(reverseLookUp(dict1, 18))
print(reverseLookUp(dict1, 22))
```
